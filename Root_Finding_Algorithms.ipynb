```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**4 - x - 10

def g(x):
    return 10 / (x**3 - 1)

def h(x):
    return (x + 10) ** 0.25
```

```python
# Fixed Point Iteration Method
candidate_roots = []
x0 = 2000 # initial guess for the root
eps = 0.01 # desired accuracy of the solution (root)
n_iter = 1000000 # number of iterations

x1 = g(x0)
while np.abs(x1 - g(x0)) > eps or n_iter > 0:
    candidate_roots.append(x1)
    x0 = x1
    x1 = g(x0)
    n_iter = n_iter - 1

print(f"Root = {x1}. Number of candidate roots = {len(candidate_roots)}. Error = {f(x1)}.")
```

```python
def fixed_point_root(func, x0, eps, n_iter):
    roots = []
    x1 = func(x0)
    x0 = x1
    roots.append(x1)
    while np.abs(x1 - func(x0)) > eps and n_iter > 0:
        roots.append(x1)
        x1 = func(x0)
        x0 = x1
        n_iter = n_iter - 1
    return x1, roots
```

```python
x11, cp1 = fixed_point_root(func=h, x0=2, eps=0.01, n_iter=100000)
```

```python
x11, len(cp1)
```

```python
x12, cp2 = fixed_point_root(func=h, x0=1, eps=0.01, n_iter=100000)
x12, h(cp2[-1]) - x12, len(cp2)
```

```python
# x^2 - 4x + 4 = 0
# (x - 2)*2 = 0 => x = 2
def quadratic(x):
    return x*x - 4*x + 4

def square(x):
    return -4 / (x - 4)

x13, cp3 = fixed_point_root(func=square, x0=5, eps=0.01, n_iter=10000)
x13, square(x13), len(cp3)
```

```python
x13, cp3 = fixed_point_root(func=square, x0=5, eps=0.00001, n_iter=10000)
x13, square(x13), len(cp3)
```

```python
x = list(range(0, 10))
y = [quadratic(xi) for xi in x]
_ = plt.figure(figsize=(10, 10))
plt.plot(x, y)
plt.hlines(0, 0, 10, "k", "--")
plt.show()
```
