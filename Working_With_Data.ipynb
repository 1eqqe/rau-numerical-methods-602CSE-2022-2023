```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# f1(x) = x^2 - 4x + 4 (pendulum, x = height of the pendulum)
# f2(t) = 10 + 2*t (position of the particle, with time)
# How does this look like for x between 0 and 100
# What are the values of x for which f(x) is maximum? What are the values of x for which f(x) is minimum?
# What are the values of x for which f(x) = 0
```

```python
# For f2
# (t = 0, f2(0) = 10)
# (t = 1, f2(1) = 12)
# (t = 2, f2(2) = 14)
# ...


def f2(t):
    result = 10 + 2 * t
    return result
f2(0), f2(1), f2(2)
```

```python
t = np.linspace(0, 100, 100)
t
```

```python
y = []
for ti in t:
    yi = f2(ti)
    y.append(yi)
y = np.array(y)
y
```

```python
fig = plt.figure(figsize=(20, 10))
plt.scatter(t, y, s=50)
plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")
plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))
plt.scatter(t, y, s=50)
plt.plot(t, y, "k")
plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")
plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))
plt.bar(t, y)
plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")
plt.show()
```

```python
def f1(x):
    result = x*x - 4*x + 4
    return result
x = np.linspace(-100, 100, 50)
y = []
for xi in x:
    yi = f1(xi)
    y.append(yi)
y = np.array(y)
x, y
```

```python
fig = plt.figure(figsize=(20, 10))
plt.scatter(x, y, s=50)
plt.xlabel("Time (s)")
plt.ylabel("Height (m)")
plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))
plt.scatter(x, y, s=50)
plt.plot(x, y, "k")
plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")
plt.show()
```

```python
def f3(x):
    result = np.sin(5*x) + 10
    return result
x = np.linspace(-np.pi, np.pi, 500)
y = []
for xi in x:
    yi = f3(xi)
    y.append(yi)
y = np.array(y)
fig = plt.figure(figsize=(20, 10))
plt.scatter(x, y, s=50)
plt.plot(x, y, "k")
plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")
plt.show()
```
